```python
#### Read the file


with open('example.txt', 'r') as file:
  content = file.read()
  print(content)
```

```python
## Read a file line by line 

with open('example.txt', 'r') as f:
  for line in f:
    print(line.strip())
    print("*"*10)
```

```python
### Writing a file 

with open('example.txt', 'w') as f:
  f.write('Helow world!\n')
  f.write('This is a new line\n\n')
```

```python
#### Write a file without over writing

with open('example.txt', 'a') as f:
  f.write('append is taking place\n')
  f.write('second append is taking place\n')
```

```python
#write list of lines
lines=['This is my first line\n',
       'second line\n',
       'third line\n']
```

```python
with open('example.txt', 'a') as f:
  f.writelines(lines)
```

```python
########### copy a file content to a new file 

with open('example.txt', 'r') as f: 
  content = f.read()
  f.close()
  print(content)
  
```

```python
with open('destination.txt', 'w') as f:
  f.write(content)
  f.close()
```

```python
#### make a new file which will add the count of char 
c=[]
with open('example.txt', 'r') as f:
  for l in f:
    c.append(f"{len(l)}  : {l}  ")
print(c)
with open('destination.txt', 'w') as ff:
  ff.writelines(c)
  
```
