```python
query="What is ML, and tell me about the it's usecasese, what is the overfitting in it explain"
```

```python
decopose_q= ChatPromptTemplate.from_template("""Given question 
                                             break it into simple sub-questions only 
                                             Question : {question}

                                             Sub-question :
                                             
                                             Instruction:
                                              - Don't explain anything like "Here are the simple sub-questions"
                                             """)
```

```python
from langchain_core.output_parsers import *
decompose_chain = decopose_q | llm | StrOutputParser()
```

```python
sub_questions=decompose_chain.invoke({"question":query})
sub_questions
```

```python
sub_questions_list = sub_questions.strip().split("\n")
sub_questions_list
```

```python
all_response=[]
for doc in sub_questions_list:
  all_response.extend(multi_ret.invoke(doc))

all_response
  
```

```python
# remove duplicate

unique_doc = list({
  doc.page_content : doc
  for doc in all_response
}.values())

unique_doc[0].page_content
```

```python
stuf_chain_p= ChatPromptTemplate.from_messages(
  [
    ('system', """You are a helpful assiatanat give ans based on the context relevance to the user question.
     
    Context: {context}
     """),
     ('human', "{input}")
  ]
)

stuf_chain=create_stuff_documents_chain(llm=llm, prompt=stuf_chain_p)
```

```python
rag_chain = create_retrieval_chain(retriever=multi_ret, combine_docs_chain=stuf_chain)
rag_chain
```

```python
final_doc = []

for doc in unique_doc:
  final_doc.extend([doc.page_content])

final_doc="\n".join(final_doc)
final_doc
```

```python
ans=rag_chain.invoke({"context":final_doc, "input":query})
print(ans['answer'])
```
