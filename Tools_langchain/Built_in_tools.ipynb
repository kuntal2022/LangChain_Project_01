## 1.DuckduckGoSearch

```python
# DuckDuckGoSearch

from langchain_community.tools import DuckDuckGoSearchRun

#Search tool 
search_tool = DuckDuckGoSearchRun()
result= search_tool.invoke('What is the election new in westbengal')

all_news=result.split('.')

for news in all_news:
  print(news)

```

## 5. SQLDatabaseQueryTool
> Runs SQL query

```python
# Create the sql table 
import sqlite3
from langchain_community.utilities import SQLDatabase

#create the db
db = SQLDatabase.from_uri("sqlite:///example.db")

#connect the db
conn = sqlite3.connect("example.db")

cursor = conn.cursor()
cursor.execute("""
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY,
        name TEXT,
        age INTEGER
    )
""")

# Clear existing data to avoid duplicates on re-run
cursor.execute("DELETE FROM users")

cursor.executemany("INSERT INTO users (name, age) VALUES (?, ?)", [
    ("Prem", 25),
    ("Rahul", 30),
    ("Priya", 22)
])
conn.commit()
conn.close()
```

```python

from langchain_community.tools import QuerySQLDataBaseTool


sql_tool = QuerySQLDataBaseTool(db=db)

result = sql_tool.invoke("SELECT * FROM users where age >=25")
print(result)
```

## 6. Write a file reading a file


```python
from langchain_community.tools import WriteFileTool, ReadFileTool

# Write tool 
writer=WriteFileTool()
writer.invoke({
  "file_path" : 'write_tool.txt',
  "text" : "My First tool writer is finnaly made"
})


#Read tool 
reader = ReadFileTool()
result = reader.invoke({"file_path":"write_tool.txt" })
print(result)
```
